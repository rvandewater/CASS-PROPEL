```python
import math
from pathlib import Path
import pandas as pd
```

```python
log_dir = Path('/Users/bjarne.pfitzner/Downloads/feature_importances_corrected_')
metric = 'prc_auc'
```

```python
runs = []
for run_instance in log_dir.iterdir():
    if run_instance.is_dir():
        iteration_result = Path(run_instance / "data_frames" / f"{metric}.csv")
        if iteration_result.exists():
            result = pd.read_csv(open(iteration_result), index_col=0)
            result['seed'] = str(run_instance).split('_')[-1]
            result.set_index('seed', append=True, inplace=True)
            # result = result.apply(lambda x: math.pow(x,e))
            runs.append(result)
all_runs = pd.concat(runs)
```

```python
cleaned_runs = all_runs[all_runs.index.get_level_values(0) != 'DummyClassifier']
```

```python
cleaned_runs
```

```python
aggregated_seeds_mean = all_runs.groupby(level=1).mean()
aggregated_seeds_mean = aggregated_seeds_mean.add_suffix("_mean")

aggregated_seeds_std = all_runs.groupby(level=1).std()
aggregated_seeds_std = aggregated_seeds_std.add_suffix("_std")
```

```python
aggregated_seeds = pd.concat([aggregated_seeds_mean, aggregated_seeds_std], axis=1)
```

```python
aggregated_seeds.sort_values(by='30d_mortality_mean', ascending=False)
```
